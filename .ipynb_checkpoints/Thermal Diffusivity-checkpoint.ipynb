# Thermal Diffusivity

In this experiment we calculated the thermal diffusivity of PEEK (polyetheretherketone) by moving a cylindrical sample of PEEK between water kept at $0^{o}C$ and $80^{o}C$ and back again every 90 seconds for 30 minutes. This means that the external temperature should be a square wave.

After a transient period, we expect the internal temperature to be sinosoidal in nature

Initialising the python interpreter:

```python
import numpy as np
from scipy.optimize import curve_fit
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

Reading in the times, internal tmperature and external temperature:

```python
#Reading in csv file
df = pd.read_csv('1.5minsexp2_final.csv', header=None)

#Seperating columns of csv fles
times = df[df.columns[0]]
int_temp = df[df.columns[1]]-0.147327586
ext_temp = df[df.columns[2]]-1.088890234

#Converting to numpy arrays for easier data analysis
np_times = times.to_numpy()
np_int_temp = int_temp.to_numpy()
np_ext_temp = ext_temp.to_numpy()

#Getting rid of data in the transient period (first 50%)
time_after_transients = np_times[int(0.5 * len(times)):]
int_temp_after_transients = np_int_temp[int(0.5 * len(times)):]
ext_temp_after_transients = np_ext_temp[int(0.5 * len(times)):]
```

Defining the expected wave for the internal temperature after the transient period. As this is sinosoidal, the wave takes the form $Asin(ωT + Φ) + B$

```python
def wave_fit(time, a, omega, phi, b):
    return a*np.sin(omega*time + phi) + b
```

Using the curevefit library, the data is then fitted to the curve

```python
initial_guess = [5,2*np.pi/200, -np.pi/2,39] #Decided after looking at graph of points

a_fit, cov = curve_fit(wave_fit,time_after_transients,int_temp_after_transients, p0=initial_guess)

#Obtaining the line of best fit and goodness of fit
fitted_curve = wave_fit(time_after_transients, a_fit[0], a_fit[1], a_fit[2], a_fit[3])
r_2_score = r2_score(int_temp_after_transients,fitted_curve)


#Plotting the graph
plt.figure()
plt.plot(time_after_transients, int_temp_after_transients, "mo", label = 'Raw Data')
plt.plot(time_after_transients,fitted_curve, label = f'{a_fit[0]:.3f}sin({a_fit[1]:.2f}t + {a_fit[2]:.2f}) +{a_fit[3]:.2f}')
plt.xlabel("time (s)")
plt.ylabel("internal temperature (°C)")
plt.title("Periodic Change in Temperature")
plt.figtext(0.15,0.83,r'The $r^{2}$ score of this fit is ' + f'{r_2_score:.4f}', fontsize = 10)

# displaying the plot
plt.legend()
plt.show()
```

As can be seen from the graph, our $r^{2}$ value is 0.9728, which is quite close to 1. This suggests that our model is quite accurate at describing the system. From this graph, we can see that the phase of the sin wave is -6.87 and the amplitude is 4.968.

From theory, we can say that $B=\frac{2ΔT}{π|M_{o}(x)|}$ where $M_{o}(x)$ is a Kelvin function and $x=a\sqrt{\frac{2π}{τD}}$

B is the peak-to-peak amplitude.
ΔT is the change in temperature.
$τ$ is the time period of 1 oscillation (180 seconds)
a is the radius of the cylinder

This can be rearranged to give
$|M_{o}(x)|=\frac{4ΔT}{Bπ}$

```python
M_peak = (4*80)/(2*a_fit[0]*np.pi)
print(M_peak)
```

Consulting a table of values for $M_{o}(x)$ [1], the value of $x$ is 5.81 

From theory we can also state that the phase shift is given by:
$φ=arg(M_{o}(x))$

To find the phase shift, we need to find the phase of the square wave. To do this we create a the expected pattern we expect the square wave to follow

```python
def square_wave_fit(time, omega, phi, b):
    return 40*np.sign(np.sin(omega*time + phi)) + b
```

Using the curevefit library, the data is, once again, then fitted to the curve

```python
initial_guess2 = [2*np.pi/180, np.pi,39]
b_fit, _ = curve_fit(square_wave_fit,time_after_transients,ext_temp_after_transients, p0=initial_guess2)


fitted_curve2 = square_wave_fit(time_after_transients, b_fit[0], b_fit[1], b_fit[2])
sin_wave = wave_fit(time_after_transients, -40, b_fit[0], b_fit[1], b_fit[2])


temp = 40*np.sin(0.03488888888888889*time_after_transients+3.14)+39.82

#Plotting Graph       
plt.figure()
plt.plot(time_after_transients, ext_temp_after_transients, "mo", label = 'Raw Data')
plt.plot(time_after_transients,fitted_curve2, label = f'Square Wave')
plt.figtext(0.728,0.41,f'Phase = {b_fit[1]:.2f}')

# setting axis labels & figures 
plt.xlabel("time (s)")
plt.ylabel("enternal temperature (°C)")
plt.title("Periodic Change in Temperature")

# displaying the plot
plt.legend()
plt.show()
```

The phase shift is the difference in phase between the internal and external temperature. It should be a number between $0$ and $2π$

```python
phase_shift = b_fit[1]-a_fit[2]

while phase_shift > 2*np.pi:
    phase_shift -= 2*np.pi
while phase_shift < 0:
    phase_shift += 2*np.pi
    
print(phase_shift)
```

Using the same Table, this corresponds to a values of 5.86 for x

Rearranging the formula for x, $D=\frac{2π}{(\frac{x}{a})^2{τ}}$
The uncertainty is given by $δD=(\frac{2ω}{x^2}a)(δD)$

```python
#Calculating Diffusivity for each method
D_amp = 2*np.pi/(180*(5.81/12.85)**2)
D_phase = 2*np.pi/(180*(5.86/12.85)**2)

#Calculating uncertainty for each method
Delta_amp = (2*2*np.pi*12.85/(5.81**2))*0.02
Delta_phase = (2*2*np.pi*12.85/(5.86**2))*0.02

#Calculating weighted mean and uncertainty
D_weighted= ((D_amp/(Delta_amp**2))+(D_phase/(Delta_phase**2)))/((1/(Delta_amp**2))+(1/(Delta_phase**2)))
D_uncertainty = 1/np.sqrt((1/(Delta_amp**2))+(1/(Delta_phase**2)))

print(f'The diffusivity of PEEK is therefore given by D = {D_weighted:.3f} ± {D_uncertainty:.3f}')
```

The generally accepted value of the diffusivity of PEEK is 0.206$mm^2s^{-1}$ [2], which falls within 1 standard deviation of our result. This means that our findings are constitent with theory
